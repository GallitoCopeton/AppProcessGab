```python
#Read qrs excel
qrs_sz_excel = pd.ExcelFile("../data/QRSZ.xlsx")
qrs_sz_df = qrs_sz_excel.parse('QRS')

#list of qrs to read and compare
qrs = list(qrs_sz_df['QR'])
listQrs = []
for qr in qrs:
    
    cursor_register = db.registerstotals.find({'qrCode': str(qr)})
    if(cursor_register): 
        for doc in cursor_register:
            resultDict = dict()
            resultDict['QR'] = doc['qrCode']
            for marker in doc['marker']:
                resultDict[marker['name']] = marker['result']
            listQrs.append(resultDict)
    else: 
        print("No records!")
#Cursor to DataFrame
headers = ['QR','ESAT6', 'CFP10', 'RV1681', 'P24',
          'P26','Control']
result_df = pd.DataFrame.from_dict(listQrs)
result_df = result_df[headers]
result_df
```
