```python
#Leer excel pruebas hematocrito bajo
casos_hema_excel = pd.ExcelFile("../../Casos/erroresCapturaEjemplos.xlsx")
casos_hema_df = casos_hema_excel.parse('Anemia')
casos_hema_df.head()
```

```python
def colorStats(img, hState, qr):
    markers = ["ESAT6","CFP10", "RV1681", "P24", "H2", "Control"]
    list_color_stats = []
    statsMeasures = ["Mean", "Median", "STD", "P25"]

    if(not isinstance(img, str)):
        indXMs = inA.resizeAll(processing.cropXMarkers(img),90)  
        
        if(isinstance(indXMs,str)):
            print(indXMs)
        else:
            #Estadisticas x marcador
            for i,m in enumerate(indXMs):
                statsXMarker = dict()
                statsH , statsS, statsV = inA.controlStats(m)
                statsXMarker['QR'] = qr
                statsXMarker['Anemia'] = hState
                statsXMarker['XMarker'] = markers[i]
                for j,h in enumerate(statsH):
                    #Guardar resultados en diccionario
                    statsXMarker['H_' + statsMeasures[j]] = h
                    statsXMarker['S_' + statsMeasures[j]] = statsS[j]
                    statsXMarker['V_' + statsMeasures[j]] = statsV[j]
                    
                list_color_stats.append(statsXMarker)

    return list_color_stats
```

```python
#Procesar lista de qrs
qrs = casos_hema_df.QR
list_color_stats = []

for i,qr in enumerate(qrs): 
    images = rI.readImage(str(qr), array = True)
    if(not isinstance(images, str)):
        for img in images:
            #Recortar marcadores
            list_color_stats += colorStats(img, casos_hema_df.iloc[i]['Anemia'], qr)

color_stats_df = pd.DataFrame.from_dict(list_color_stats)
color_stats_df.head()

#print(statsXMarker)
```
